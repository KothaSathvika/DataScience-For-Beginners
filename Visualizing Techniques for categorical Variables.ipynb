# Visualizing Techniques for categorical Variables

Frequency Distribution Table:

```python
import pandas as pd
import numpy as np
```

```python
# initialize data of lists.
data = {'CarBrand' : ['Audi', 'Mercedes', 'BMW'],
       'Frequency' : [124,113,98]}
# Create DataFrame
df = pd.DataFrame(data)
# Print the output.
df
```

Bar Chart:

```python
from matplotlib import pyplot as plt
# using a function df.plot.bar()
#specifying to take x ticks as CarBrand
df.plot(kind="bar",x='CarBrand')
plt.xlabel("Car Brand")
plt.ylabel("Frequency")
```

Pie Chart:

Pie Chart is also used for the visual representation of categorical variables. But pie chart uses the relative frequency. In pandas we have dataframe.plot.pie() function to represent the pie chart so it automatically calculates the relative frequency of the specific category.

```python
# using a function df.plot.pie()
df.plot(kind = "pie",y="Frequency", autopct='%.2f',labels=df.CarBrand)
```

Pareto Diagrams:

```python
# Add cumulative percentage column
df["cum_percentage"] = round(df["Frequency"].cumsum()/df["Frequency"].sum()*100,2)
df
```

```python
from matplotlib.ticker import PercentFormatter

# Set figure and axis
fig, ax = plt.subplots(figsize=(22,10))

# Plot bars (i.e. frequencies)
ax.bar(df.CarBrand, df["Frequency"])
ax.set_title("Pareto Chart")
ax.set_xlabel("Car Brands")
ax.set_ylabel("Frequency");

# Second y axis (i.e. cumulative percentage)
ax2 = ax.twinx()
ax2.plot(df.CarBrand, df["cum_percentage"], color="red", marker="D", ms=7)
ax2.axhline(80, color="orange", linestyle="dashed")
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylabel("Cumulative Percentage");
```
