# Visualizing Techniques for Numerical Data

Frequency Distribution Table with intervals:

```python
import pandas as pd
import numpy as np
```

```python
# initialize data
data = {'DataSet': np.random.randint(1, 100, size=25),
       'Frequency': [1 for _i in range(25)]}
# Create DataFrame
df = pd.DataFrame(data)
#Print Output
df
```

```python
#using pd.cut for creating intervals
df['DataSet_bins'] = pd.cut(x=df['DataSet'], bins=[0, 25, 50, 75, 100])
```

```python
#grouping by DataSet_bins 
df = df.groupby(df['DataSet_bins']).count()
df.drop(['DataSet'],axis=1, inplace = True)
```

```python
df['DataSet_bins'] = df.index
#Resetting the index to normal
df.reset_index(drop=True, inplace=True)
df
```

```python
#changing the columns 
df = df[['DataSet_bins', 'Frequency']]
df
```

Histogram Charts

```python
# initialize data
data = {'DataSet': np.random.randint(1, 100, size=25),
       'Frequency': [1 for _i in range(25)]}
# Create DataFrame
df1 = pd.DataFrame(data)
#Print Output
df1
```

```python
from matplotlib import pyplot as plt
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(df1, bins =[0, 50, 100])
 
# Show plot
plt.show()
```
